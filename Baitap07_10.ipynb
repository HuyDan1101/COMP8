```python
#W4A8
n = int(input('Nhập số nguyên dương n : '))
def so_dao_nguoc(a):
  a_daonguoc = 0
  while a > 0:
    du = a % 10
    a_daonguoc = a_daonguoc * 10 + du
    a = a // 10
  return a_daonguoc
n = n + so_dao_nguoc(n)
so_lan = 1
while n - so_dao_nguoc(n) != 0:
  n = n + so_dao_nguoc(n)
  so_lan += 1
print(so_lan, n)
```

```python
#W4A10
n = int(input('Nhập số nguyên dương n : '))
def collatz(a):
  L = 1
  while a != 1:
    if a % 2 == 0:
      a /= 2
      L += 1
    else:
      a = a * 3 + 1
      L += 1
  return L
max_collatz = collatz(1)
max = 1
for x in range(1, n + 1):
  if collatz(x) > max_collatz:
    max_collatz = collatz(x)
    max = x
  else:
    max_collatz = max_collatz
    max = max
print(max, max_collatz)

```

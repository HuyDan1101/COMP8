```python
#Chương trình nhập vào một số nguyên n và in ra các chữ số đảo ngược nhau
n = int(input("Nhập một số nguyên: "))
n_daonguoc = 0
while n > 0:
  phandu = n % 10
  n_daonguoc = n_daonguoc * 10 + phandu
  n = n // 10
print('So dao nguoc la: ', n_daonguoc)
```

```python
#Triangle
a = int(input('Nhập độ dài cạnh thứ nhất của tam giác: '))
b = int(input('Nhập độ dài cạnh thứ hai của tam giác: '))
c = int(input('Nhập độ dài cạnh thứ ba của tam giác: '))
if a + b > c and a + c > b and b + c > a:
  print('Yes')
else:
  print('No')
```

```python
#Largest number
a = int(input('Số nguyên thứ nhất: '))
b = int(input('Số nguyên thứ hai: '))
c = int(input('Số nguyên thứ ba: '))
d = int(input('Số nguyên thứ tư: '))
day = (a,b,c,d)
max = a
for i in day:
  if max < i:
    max = i
print(max)
```

```python
#Phân loại tam giác
a,b,c = map(int, input('Nhập độ dài 3 cạnh tam giác: ').split())
if a + b <= c or a + c <= b or b + c <= a:
  print('Không phải tam giác')
elif a == b == c:
  print('Tam giác đều')
elif a == b > c / 2 or b == c > a / 2 or a == c > b / 2:
  print('Tam giác cân')
else:
  print('Tam giác thường')
```

```python
#Tính tiền điện
sodien = int(input('Nhập số kWh điện tiêu thụ: '))
if sodien <= 50:
  tiendien = sodien * 1500
elif 50 < sodien <= 100:
  tiendien = 50 * 1500 + ( sodien - 50 ) * 2000
else:
  tiendien = 50 * 1500 + 50 * 2000 + ( sodien - 100 ) * 3000
print(tiendien)
```

```python
#Giải phương trình bậc nhất
print('Phương trình ax + b')
a = float(input('Nhập số thực a: '))
b = float(input('Nhập số thực b: '))
if a == b == 0:
  print('Vô số nghiệm')
elif a == 0 and b != 0:
  print('Vô nghiệm')
else:
  print('%.2f'%((- b) / a))

```
